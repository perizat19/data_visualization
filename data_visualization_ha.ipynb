# Line Charts

```python
import matplotlib.pyplot as plt
import seaborn as sns

years = range(2020, 2025)
yield_apples = (0.52, 0.78, 0.84, 0.83, 0.91)
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield apples(tones per hectare)')
plt.show()
```

```python
yield_cherry = (0.76, 0.68, 0.79, 0.80, 0.95)
years = range(2020, 2025)
yield_apples = (0.52, 0.78, 0.84, 0.83, 0.91)
plt.plot(years, yield_cherry)
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield apples(tones per hectare)')
plt.legend(['Cherry', 'Apples'])
plt.title('Crop yields in Kanto')
plt.show()
```

```python
sns.set_style('whitegrid')
coconut = (0.76, 0.81, 0.85, 0.89, 0.95)
cotton = (1.00, 0.98, 0.84, 0.83, 0.80)
year = range(2020, 2025)

plt.plot(year, coconut, marker='o')
plt.plot(year, cotton, marker='x')

plt.xlabel('Year')
plt.ylabel('Crop Yields (tons per hectare)')

plt.legend(['Coconut', 'Cotton'])
plt.show()
```

```python
sns.set_style('darkgrid')
cotton = (1.00, 0.98, 0.84, 0.83, 0.80)
plt.plot(year, cotton)
plt.figure(figsize=(20,14))
plt.show()
```

# Bar Graphs

# Histograms

```python
iris_df = sns.load_dataset('iris')
iris_df
```

```python
sns.barplot(x='species', y='sepal_width', data=iris_df)
plt.show()
```

```python
sns.barplot(x='species', y='sepal_width', hue='petal_width', data=iris_df)
plt.show()

```

```python
iris_df.sepal_width
```

```python
plt.hist(iris_df.sepal_width)
plt.show()
```

```python
plt.hist(iris_df.sepal_length, bins=5)
plt.show()
```

```python
import numpy as np
plt.hist(iris_df.sepal_length, bins=np.arange(4, 8, 0.25))
plt.show()
```

```python
setosa_df = iris_df[iris_df.species == 'setosa']
virginica_df = iris_df[iris_df.species == 'virginica']
versicolor_df = iris_df[iris_df.species == 'versicolor']

plt.hist(setosa_df.sepal_width, alpha=0.5, bins=np.arange(2, 5, 0.25))
plt.hist(versicolor_df.sepal_width, alpha=0.5, bins=np.arange(2, 5, 0.25))
plt.show()
```

```python
plt.hist(versicolor_df.petal_length, alpha=0.4, bins=np.arange(3, 8, 0.25))
plt.hist(virginica_df.petal_length, alpha=0.4, bins=np.arange(3, 8, 0.25))
plt.show()
```

```python
plt.hist(versicolor_df.petal_width, alpha=0.4)
plt.show()
```

```python
plt.hist([versicolor_df.sepal_width, virginica_df.sepal_width, setosa_df.sepal_width], bins=np.arange(2, 5, 0.25), stacked=True)
plt.show()
```

```python
plt.hist(setosa_df.petal_width, alpha=0.4, bins=np.arange(0, 3, 0.25))
plt.hist(setosa_df.petal_length, alpha=0.4, bins=np.arange(0, 3, 0.25))
plt.show()
```

```python
plt.hist([setosa_df.sepal_length, versicolor_df.sepal_length, virginica_df.sepal_length], bins=np.arange(4, 8, 0.5), stacked=True)
plt.legend(['Setosa', 'Versocolor', 'Virginica'])
```

# Pie Chart

```python
import numpy as np


x = np.array([25, 25, 10, 40])
plt.pie(x)
plt.show()
```

Labels 

To add labels to the pie chart, I will use labels parameter.

```python
x = np.array([25, 25, 10, 40])
mylabels = ['Sugar', 'Water', 'Oil', 'Butter']
plt.pie(x, labels=mylabels)
plt.show()
```

Explode

```python
x = np.array([25, 25, 10, 40])
mylabels = ['Sugar', 'Water', 'Oil', 'Butter']
myexplode = [0.1, 0, 0, 0]
plt.pie(x, labels=mylabels, explode=myexplode)
plt.show()
```

Shadow

```python
x = np.array([25, 25, 10, 40])
mylabels = ['Sugar', 'Water', 'Oil', 'Butter']
myexplode = [0.1, 0, 0, 0]
plt.pie(x, labels=mylabels, explode=myexplode, shadow=True)
plt.show()
```

Colors

```python
x = np.array([25, 25, 10, 40])
mylabels = ['Sugar', 'Water', 'Oil', 'Butter']
mycolors = ['pink', 'purple', 'c', 'yellow']
plt.pie(x, labels=mylabels, colors=mycolors)
plt.show()
```

Legend

I use .legend() function to explain the meaning of each wedge in my pie chart

```python
x = np.array([25, 25, 10, 40])
mylabels = ['Sugar', 'Water', 'Oil', 'Butter']
mycolors = ['pink', 'purple', 'c', 'yellow']
plt.pie(x, labels=mylabels, colors=mycolors)
plt.legend()
plt.show()
```
